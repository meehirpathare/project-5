```python
c_list = make_color_list('all_colors.txt')
```

```python
step = 10e5
```

```python
low = 0
high = step

lowers = []
uppers = []
while low < len(c_list):
    lowers.append(low)
    low += step

while high < len(c_list):
    uppers.append(int('{:06.0f}'.format(high -1)))
    #print(type('{:.0f}'.format(high)))
    high +=  step
   
uppers.pop(-1)
uppers.append(len(c_list))

lims = list(zip(lowers, uppers))
for i, j in lims:
    print("make_color_map(c_list, palette, data_path + 'cmap_{}_to_{}.txt')".format(i, j))   
```
