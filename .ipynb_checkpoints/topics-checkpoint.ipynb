```python
def merge_cols(df, col1, col2, new_col_name,drop_cols=True):
    df[new_col_name] = df[col1] + df[col2]
    
    if drop_cols:
        df.drop(columns=[col1, col2], inplace=True)
        return ("merged {}[{}] and {}[{}] into {}[{}] and dropped columns=[{}, {}]".format(df, col1, df, col2, df, new_col_name, col1, col2))
    else:
        return ("merged {}[{}] and {}[{}] into {}[{}] and dropped columns=[{}, {}]".format(df, col1, df, col2, df, new_col_name, col1, col2))
```
