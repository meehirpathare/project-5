```python
from colorthief import ColorThief
import numpy as np 
from helpers import display_color, get_channel
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
from skimage import io
%matplotlib inline
```

```python
img = mpimg.imread('test_image.jpg').astype('uint8')
color_thief = ColorThief('test_image.jpg')
```

```python
# get the dominant color
dominant_color = color_thief.get_color(quality=10)
# build a color palette
pal = color_thief.get_palette(color_count=21)
```

```python
len(pal)
```

```python
color_names = []
for i in range(1, 21):
    color_names.append('Color {}'.format(i))
print(color_names)
```

```python
palette = {}
for i in range(0, len(pal)):
    palette["Color {}".format(i + 1)] = pal[i]
print(palette)
color_list = np.array([x for x in palette.values()], dtype=np.uint8)
```

```python
img = mpimg.imread('test_image.jpg') 
img = img.astype('uint8')
        
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, nrows=1)
fig.set_figheight(15)
fig.set_figwidth(15)

ax1.imshow(get_color_channel(img, 'red'), cmap='Reds')
ax1.set_axis_off()

ax2.imshow(get_color_channel(img, 'green'), cmap='Greens')
ax2.set_axis_off()

ax3.imshow(get_color_channel(img, 'blue'), cmap='Blues')
ax3.set_axis_off()

fig.set_axis_off()
plt.show()
```
