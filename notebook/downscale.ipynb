```python
data_path = "D:\\METIS\\data\\"
image_path = "D:\\METIS\\images\\"
resized_path = "D:\\METIS\\downscaled\\"
```

```python
list_of_images = os.listdir(image_path)
len(list_of_images)
```

```python
list_of_images[4]
```

```python
downsized_imgs = []
start = datetime.now()
```

```python
for i,name in enumerate(list_of_images):
    if i % 1000 == 0:
        with open(data_path + "downsized_imgs.json", "w") as write_file:
            json.dump(downsized_imgs, write_file)  

        print("{}: {} elapsed".format(i, datetime.now() - start))
    
    try:
        pil_img = Image.open(image_path + name)
        img = pil_img.copy()
        img.thumbnail((224, 224))
        image = np.asanyarray(img)
        filename = "resized_" + name.replace('.jpg', '', 1)
        imsave(fname= resized_path + filename, arr=image)
        
        downsized_imgs.append({
                                'filename': filename
                                })
    except:
        pass
```
