```python
roles = []
names = []
genders = []

for response in data:
    if response['constituents'] is None:
        continue
    else:
        roles.append(response['constituents'][0]['role'])
        names.append(response['constituents'][0]['name'])
        genders.append(response['constituents'][0]['gender'])
```

```python
role_counts = {}
for role in set(roles):
    role_counts[role] = roles.count(role)
```

```python
name_counts = {}
for name in set(names):
    name_counts[name] = names.count(name)
```

```python
gender_counts = {}
for gender in set(genders):
    gender_counts[gender] = genders.count(gender)
```

```python
departments = []
for response in data:
    departments.append(response['department'])
```

```python
department_counts = {}
for wing in set(departments):
    department_counts[wing] = departments.count(wing)
```

```python
df = pd.DataFrame.from_dict(data)
```

```python
df['constituents'].to_list()[0]
```

```python
def get_col_val(val, key):
    if val == None:
        return ''
    else:
        return val[0][key]
```

```python
df['constituentID'] = df['constituents'].apply(lambda x: get_col_val(x, 'constituentID'))
df['role'] = df['constituents'].apply(lambda x: get_col_val(x, 'role'))
df['person'] = df['constituents'].apply(lambda x: get_col_val(x, 'name'))
df['constituentULAN_URL'] = df['constituents'].apply(lambda x: get_col_val(x,'constituentULAN_URL'))
df['constituentWikidata_URL'] = df['constituents'].apply(lambda x: get_col_val(x,'constituentWikidata_URL'))
df['gender'] = df['constituents'].apply(lambda x: get_col_val(x, 'gender'))
```
