```python
from datetime import datetime
from get_img import get_img
import json
import os
import requests
import urllib.request
```

```python
data_path = "D:\\METIS\\data\\"
image_path = "D:\\METIS\\images\\"
```

```python
with open(data_path + 'urls.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
data[0]
```

```python
# check for duplicates
ids = []
for i in range(len(data)):
    ids.append(data[i]['object_id'])

object_ids = list(set(ids))
len(object_ids)
```

```python
images_saved = []
failed_imgs = []
```

```python
start = datetime.now()

for count, object_id in enumerate(object_ids):
    if count % 2500 == 0:
        elapsed = str(datetime.now() - start)[0:7]
        print("{}: {}".format(count, elapsed))
    
    try:
        save_data = get_img(object_id, image_path, return_meta=True)
        images_saved.append(save_data)
        
        with open(data_path + "images_saved.json", "w") as write_file:
            json.dump(images_saved, write_file)
            
    except:
        failed_imgs.append({
                            'object_id': object_id,
                            'api_url':'https://collectionapi.metmuseum.org/public/collection/v1/objects/{}'.format(object_id)
                            })
        
        with open(data_path + "failed_images.json", "w") as write_file:
            json.dump(failed_imgs, write_file)          
```
