```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
data_path = "D:\\METIS\\data\\"
```

```python
with open(data_path + 'dists.json', 'r') as f:
    data = json.load(f)
```

```python
img1_dists = []

for dist in data:
    if dist['image 1'] == '/content/images/resized_01_12_5_O1_sf.jpg':
        img1_dists.append(dist)
```

```python
dists = []
for dist in img1_dists:
    dists.append(dist['similarity'])
    
sorted_dists = np.argsort(dists)[::-1][:]
```

```python
top_10 = sorted_dists[0:10]
top_10
```

```python
top_10_images = []

for num in top_10:
    top_10_images.append(img1_dists[num])
```

```python
top_10_images
```

```python
parsed_list = []
for img in top_10_images:
    parsed_list.append({
                        'image 1': img['image 1'].replace('/content/images/', ''),
                        'image 2': img['image 2'].replace('/content/images/', ''),
                        'similarity': img['similarity']
                        })
```

```python
parsed_list[0]
```

```python
resized_path = "D:\\METIS\\downscaled\\"
```
