```python
import matplotlib.pyplot as plt
import numpy as np
from helpers import get_palette, simplify_axis, closest_on_palette
from itertools import permutations 
from datetime import datetime
import os
import json
```

```python
data_path = os.getcwd() + "/"
palette = get_palette()
```

```python
ax = np.ones(256)
```

```python
[arr, centers] = simplify_axis(ax=np.ones(256), size=3, return_arr=True, return_centers=True)
```

```python
simplified_colors = []
for i in centers:
    for j in centers:
        for k in centers:
            color = (i, j, k)
            simplified_colors.append(color)

len(simplified_colors)
```

```python
simple_cmap = {}
```

```python
start = datetime.now()
for i,color in enumerate(simplified_colors):
    
    simple_cmap[color] = closest_on_palette(color, palette)
    
    if i % 10000 == 0:
        with open(data_path + "simple_cmap.json", "w") as write_file:
            json.dump(str(simple_cmap), write_file) 
            
        print("{}: {} elapsed".format(i, datetime.now() - start))
```

```python
with open(data_path + "simple_cmap.json", "w") as write_file:
            json.dump(str(simple_cmap), write_file)
```

```python
cmap = simple_cmap.copy()
#centers[0:10], arr
```

```python
colors = [list(x) for x in arr]
```

```python
test = colors[0] + colors[0] + colors[0]
test
```

```python
perms = permutations(test, 3)
val = simple_cmap[(0,0,0)]

for p in list(perms):
    cmap[p] = val
len(cmap)   
```

```python
for i in range(1, len(colors)):
    
    lst = colors[i] + colors[i] + colors[i]
    perms = permutations(lst, 3)
    val = simple_cmap[(colors[i][1], colors[i][1], colors[i][1])]
    
    for p in list(perms):
        cmap[p] = val
    len(cmap)
```

```python
with open(data_path + "full_cmap.json", "w") as write_file:
            json.dump(str(cmap), write_file)
```
